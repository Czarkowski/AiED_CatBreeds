```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
from collections import Counter
import kagglehub
from sklearn.decomposition import PCA
import numpy as np
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder
from PIL import Image, ImageEnhance
import random
from pathlib import Path
import seaborn as sns
```

```python
data = np.load("dsprites-dataset-master/dsprites_ndarray_co1sh3sc6or40x32y32_64x64.npz", allow_pickle=True)

print(data.files)
```

```python
imgs = data["imgs"]
latents_values = data["latents_values"]
latents_classes = data["latents_classes"]
metadata = data["metadata"].item()

print("Shape imgs:", imgs.shape)
print("Latents:", latents_classes.shape)
```

```python
latent_names = metadata["latents_names"]
print("Nazwy latentów:", latent_names)

df_latents = pd.DataFrame(latents_classes, columns=latent_names)
print(df_latents.describe())
```

```python
shape_counts = df_latents["shape"].value_counts()

plt.figure(figsize=(6,4))
shape_counts.plot(kind="bar")
plt.title("Rozkład klas - shape")
plt.xlabel("Shape")
plt.ylabel("Liczba próbek")
plt.show()
```

```python
plt.figure(figsize=(6,6))

for i in range(9):
    plt.subplot(3,3,i+1)
    plt.imshow(imgs[i], cmap="gray")
    plt.title(f"{latents_classes[i]}")
    plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
df_latents.hist(figsize=(10,8))
plt.suptitle("Rozkład latentów")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df_latents.corr(), annot=True, cmap="coolwarm")
plt.title("Korelacja latentów")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df_latents["shape"], y=df_latents["scale"])
plt.title("Scale vs Shape")
plt.show()
```
